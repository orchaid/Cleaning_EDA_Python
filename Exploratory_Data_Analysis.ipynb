### Understanding the Data

```python
df.select_dtypes('number') # 'float', 'object' to filter fast based on columns data type
```

```python
df.sort_values('2022 Population', ascending=False).head() # biggest countries in terms of population 2022
```

```python
df.corr(numeric_only=True) # calculates the correlation between numeric columns
```

```python
sns.heatmap(df.corr(numeric_only=True),annot= True) # show the corr with more glamor

plt.rcParams['figure.figsize' ]= (20,7) # tp access the parameters of the figure
plt.show()
```

### Visulaizing Continents Populations

```python
df.groupby('Continent').mean('Growth Rate').sort_values(by= '2022 Population', ascending=False) # specifing a neumeric column inside mean so that it doesn't calculate the mean of non-numeric values
```

```python
df_grouped = df.groupby('Continent')[df.columns[::-1][5:12]].mean().sort_values(by= '2022 Population', ascending=False) 
df_grouped.transpose()
```

```python
sns.lineplot(data=df_grouped.transpose()) # Asia's population trend is the first, and in the second place africa
# I can dig deeper to see which countries are the most populated in each continent
```
