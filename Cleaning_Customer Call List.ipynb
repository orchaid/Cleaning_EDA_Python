### Dropping NaN

Don't use `.replace('[NaN,Na,N/a]','')` as Square brackets [] define a character class, meaning it looks for any of the characters `N`, `a`, `/`, or `,` individually.
- Use `.replace['','']`

`replace()` does not handle `NaN` unless it's explicitly converted to a string.
If the DataFrame contains actual `NaN` values (not strings), use:
- `df.replace([np.nan, 'Na', 'N/a'], '')` or remove NaN by `fillna('')` then user the first replace and it will work.

```python
import numpy as np

df = df.replace([np.NaN,'Na','N/a'],'', regex=True)
df
```
